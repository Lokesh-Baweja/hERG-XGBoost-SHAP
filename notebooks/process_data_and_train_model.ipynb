```python
import sys
import os

sys.path.append(os.path.abspath(".."))
from src.data_loader import load_dataset
from src.data_preprocessing import generate_fingerprints
from src.train import train_models
from src.evaluate import evaluate_models

# Load data
train_smiles, train_y, test_smiles, test_y = load_dataset("../data/processed/training.csv", "../data/processed/test.csv")

# Generate features
X_train = generate_fingerprints(train_smiles)
X_test = generate_fingerprints(test_smiles)

# Train & tune models
rf_best, xgb_best = train_models(X_train, train_y)

# Evaluate models
results = evaluate_models(rf_best, xgb_best, X_test, test_y)
```
